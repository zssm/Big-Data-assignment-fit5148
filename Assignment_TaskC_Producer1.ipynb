```python
import csv

# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt

data=[]
with open('climate_streaming.csv') as f:
    reader = csv.reader(f)
    
    count=0
    header1=[]
    for row in reader:
        temp=[]
        if count==0:
            
            count=1
            continue
       
         

        else:
            datetime = str(dt.datetime.now().strftime("%X"))
        
            d={'Latitude':float(row[0]),
               'Longitude':float(row[1]),
               'Air Temperature (Celcius)':int(row[2]),
               'Relative Humidity':float(row[3]),
                'Windspeed (knots)':float(row[4]),
                'Max Wind Speed':float(row[5]),
                'Precipitation':str(row[6]),
                'Sender ID':0,
                'Hotspot':[]
               
                
              }
            data.append(d)


            

def publish_message(producer_instance, topic_name, data):
    try:
        value_bytes = bytes(data, encoding='utf-8')
        producer_instance.send(topic_name, value=value_bytes)
        print('Message published successfully. ' + data)
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['127.0.0.1:9092'],
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = '2717318'
    print('Publishing records..')
    producer01 = connect_kafka_producer()

    
    while True:
        random_num= random.randrange(0,len(data)) 
    
        senddata=data[random_num] # randomly select data
        senddata['Created_time']=dt.datetime.now().strftime("%X")
        #genrerate  create time
        
        senddata=str(senddata) #convert to string for sending
        
        publish_message(producer01, topic, senddata) 
        sleep(5) #send every 5 seconds
```
