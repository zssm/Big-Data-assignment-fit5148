```python
#ID: 27173186 Name: Shengming Zhao
#ID:29593387 Name:Jipeng Yin
import csv
import datetime as dt
            
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random


 
data=[]
with open('hotspot_AQUA_streaming.csv') as f:
    reader = csv.reader(f)
    
    count=0
    
    for row in reader:
        
        if count==0: #skip header
           
            count=1
            continue
       
         

        else:
        
            d={'Latitude':float(row[0]),
               'Longitude':float(row[1]),
               'Confidence':int(row[2]),
               'Surface Temperature (Celcius)':int(row[3]),
                
                'Sender ID':1
            }
            data.append(d)



def publish_message(producer_instance, topic_name, data):
    try:
        value_bytes = bytes(data, encoding='utf-8')
        producer_instance.send(topic_name, value=value_bytes)
        print('Message published successfully. ' + data)
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['127.0.0.1:9092'],
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = '2717318'
    print('Publishing records..')
    producer02 = connect_kafka_producer()

        
    while True:
        random_num= random.randrange(0,len(data))
        
        senddata=data[random_num] #randomly select data
        senddata['Created_time']=dt.datetime.now().strftime("%X") # genrerate create time
        senddata['Hour']=int(senddata['Created_time'][0:2])
        senddata=str(senddata)
    
        publish_message(producer02, topic, senddata)
        sleep_number=random.randrange(10,30)
        #sleep(0.15)
        sleep(int(sleep_number))
```
