There are 2 datasets in this task, one is Hotspot, the other is climate. After checking 2 dataset, we can find that both 2 datasets have date attributes. For climate data, date has no duplicate, but hotspot could have multiple rows have the same date. Therefore, For each climate date, there could be 0 or mutiple hotspot data rows related. Therefore, we used embedded data form, store hotspot data as a attributes of climate according date.The example is shown below.The advantage of this design is that when we want to combine climate and hotspot data accroding to date, no aggregation is needed.

```python
import csv
import pymongo
from pymongo import MongoClient
from pprint import pprint

# Create a MongoClient
client = MongoClient()



db = client.fit5148_assignment_db
db.drop_collection('fit5148_collection')
collection = db.fit5148_collection

with open('hotspot_historic.csv') as f:
    reader = csv.reader(f)
    hotspot={}
    count=0
    header1=[]
    for row in reader:
        temp=[]
        if count==0:
            header1.append(row) # store header
            count=1
            continue
        else:
            
            if row[4] not in hotspot.keys():# if key not in hotspot, create a new key
                d={'Latitude': float(row[0]),
                   'Longitude': float(row[1]),
                   'Datetime': row[2],
                   'Confidence': int(row[3]),
                   'Date': row[4],
                   'Surface Temperature (Celcius)': int(row[5]),
                   'Hour':  int(row[2][11:13])
                   
                   
                  }
                
                
                
                temp.append(d)
               
                hotspot[row[4]]=temp   
            else: #key exists, update it.
                d={'Latitude': float(row[0]),
                   'Longitude': float(row[1]),
                   'Datetime': row[2],
                   'Confidence': int(row[3]),
                   'Date': row[4],
                    'Hour': int(row[2][11:13]),
                   'Surface Temperature (Celcius)': int(row[5])
                   }
                temp=hotspot[row[4]] #update existed key
                temp.append(d)
                hotspot[row[4]]=temp
        
        
        

            

with open('climate_historic.csv') as f:
    climate={}
    reader = csv.reader(f)
    header2=[]
    count=0
    for row in reader:
        if count==0:
            header2.append(row)
            count=1
            continue
        else:
            
            if row[1] in hotspot.keys():
                ahotspot=hotspot[row[1]]
                a_day={
                "Station" :int(row[0]),
                "Date" : row[1],
                "Air Temperature (Celcius)" : int(row[2]),
                    
                "Relative Humidity" : float(row[3]),
                "WindSpeed (Knots)" : float(row[4]),
                "Max Wind Speed" : float(row[5]),
                "Precipitation" : row[6],
                "Hotspot" : ahotspot # store hotspot as a attribute, store  them in a list.
                }
                
            
           
            else:
                a_day={
                "Station" : int(row[0]),
                "Date" : row[1],
                "Air Temperature (Celcius)" : int(row[2]),
                "Relative Humidity" : float(row[3]),
                "WindSpeed (Knots)" : float(row[4]),
                "Max Wind Speed" : float(row[5]),
                "Precipitation" : row[6],
                "Hotspot" : []
                }
        collection.insert_one(a_day) 


            

            
            
        
           


          
    
```

```python
# b
collection =db.fit5148_collection
result =collection.aggregate([{'$unwind': '$Hotspot'},
                                {'$match':{'$and':[{'Hotspot.Surface Temperature (Celcius)':{"$gte":65}},
                                                   {'Hotspot.Surface Temperature (Celcius)':{"$lte":100}}]}},
                                {'$project':{'Hotspot.Latitude':1,'Hotspot.Longitude':1,'Hotspot.Confidence':1,'_id':0}}])

for i in result:
    pprint(i)
```

```python
#c
collection =db.fit5148_collection

result=collection.aggregate([{'$unwind':'$Hotspot'},
                             {'$match':{'$or':[{'Date':'15/12/2017'},
                                                {'Date':'16/12/2017'}]}},
                             {'$project':{'Date':1,'Air Temperature (Celcius)':1,
                                          'Relative Humidity' :1,'Max Wind Speed' :1,
                                          'Hotspot.Surface Temperature (Celcius)':1,'_id' :0}}])

for i in result:
    pprint(i)
       
```

```python
#e
collection =db.fit5148_collection
result=collection.aggregate([{'$unwind': '$Hotspot'},
                            {'$sort': {'Hotspot.Surface Temperature (Celcius)':-1}},
                            {'$limit': 10}])
for i in result:
    pprint(i)
```

```python
# f 
collection =db.fit5148_collection
result =collection.aggregate([{'$unwind':'$Hotspot'},
                            {'$group': {
                             '_id':'$Date',
                            'count':{'$sum': 1}}}])

for i in result:
    pprint(i)
```

```python
#g 
collection =db.fit5148_collection
result=collection.aggregate([{'$unwind':'$Hotspot'},
                            {'$group': {'_id':'$Date',
                                       'average':{'$avg':'$Hotspot.Surface Temperature (Celcius)'}}}])

for i in result:
    pprint(i)
```
